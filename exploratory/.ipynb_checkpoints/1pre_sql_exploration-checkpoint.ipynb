```python
#checking for primary key
from itertools import chain, combinations

def key_options(items):
    return chain.from_iterable(combinations(items, r) for r in range(1, len(items)+1) )


# iterate over all combos of headings, excluding ID for brevity
for candidate in key_options(list(ann)[1:]):
    deduped = ann.drop_duplicates(candidate)

    if len(deduped.index) == len(ann.index):
        print (','.join(candidate))
```

```python
from itertools import chain, combinations

def powerset(iterable):
    "list(powerset([1,2,3])) --> [(), (1,), (2,), (3,), (1,2), (1,3), (2,3), (1,2,3)]"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))
```

```python

def find_primary_key(df):
    for col_sub in powerset(df.columns):
        if col_sub and len(df.groupby[list(col_sub)].unique) == 564452:
            print(col_sub)
```
